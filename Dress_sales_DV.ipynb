### Exploratory Data Analysis (EDA) ###
#### Dress Sales ####

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv('DressSales.csv')
df_3 = data
data.head()
data.info()
data.describe()
```

### Multi Analysis ###

```python
df_3.head()

```

```python
df_3 = df_3.drop(columns = ['Dress_ID'])
df_3 = df_3.dropna()
plt.figure(figsize=(20,12))
sns.boxplot(df_3)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca = pca.fit_transform(df_3)
pca
```

```python
sns.scatterplot(x = pca[:,0], y = pca[:,1])
```
